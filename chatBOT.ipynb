```python
conversation_data={
    "hello": "hi there! How can I Help you today?",
    "hi":"Hello ! hoe are you?",
    "how are you": "I'm just a bot, but I'm doing great! How about you?",
    "what is computer vision": "Computer vision is a field of AI that allows computers to see and interact",
    "what is AI":"AI stands for Artificial Intelligence,which enables machines to per",
    "what is deep learning":"Deep Learning is a subset of AI that uses neural networks",
    "bye": "goodbye! have a nice day!",
    "thank you":"you're welcome! Happy to help"
}
```

```python
def sample_chat(user_input,chat_history=[]):
  user_input_lower=user_input.lower()
  response=None
  for key in conversation_data:
    if key in user_input_lower:
      response =conversation_data[key]
      break

    if not response:
      response="sorry, i dont understand that, can you say somethimg else"


    chat_history.append(("You",user_input))
    chat_history.append(("Bot",response))

    return response, chat_history
```

```python
print("Welcome to mini coversation API")
print("type exit or quit to the end of the chat.\n")
chat_history=[]
while True:
  user_input=input("you:")

  if user_input.lower( )in ["exit","quit"]:
    print("Chat ended. GoodBye!")
    break
  bot_response,chat_history = sample_chat(user_input,chat_history)
  print("Bot:", bot_response)
```
